# **Problem 1**

Plot the total profit across the months



*   X label name = Month Number
*   Y label name = Total profit





```python
plt.rcParams['figure.figsize']=(10,5)
```

```python
data=data[['total_profit']]
data

plt.plot(data)

plt.title('Total profit across the months in a sales company')
plt.ylabel('Total profit')
plt.xlabel('month number')
plt.show()
```

# **Problem 2**
Read face cream and facewash product sales data and show it using the bar chart. The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
facecream_facewash_data=pd.read_csv("/content/drive/MyDrive/Python-Assignment/company_sales_data.csv")
facecream_facewash_data
```

```python
plt.rcParams['figure.figsize']=(10,5)
facecream_facewash= facecream_facewash_data[['facecream','facewash']]

facecream_facewash.plot(kind='bar')
plt.title('sales per month')
plt.ylabel('products')
plt.xlabel('months')
plt.show()
```

```python
plt.rcParams['figure.figsize']=(20,6)
facecream_facewash= facecream_facewash_data[['facecream','facewash','toothpaste','bathingsoap',	'shampoo','moisturizer']]

facecream_facewash.plot(kind='bar')
plt.title('sales per month')
plt.ylabel('products')
plt.xlabel('months')
plt.show()
```

# Problem 3
Display the number of units sold per month for each product using multiline plots. (i.e. separate plotline for each product ).

```python
facecream_facewash=facecream_facewash_data[['facecream','facewash','toothpaste','bathingsoap',	'shampoo','moisturizer']]
facecream_facewash.plot(kind='line')
plt.title(' The units sold per month for each product')
plt.xlabel('months')
plt.ylabel('total_units')
plt.show()
```
